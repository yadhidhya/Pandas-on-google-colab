```python
from urllib.request import urlretrieve
import pandas as pd
urlretrieve('https://raw.githubusercontent.com/the-stranger-web/jovian_Data_Analyst/refs/heads/main/italy-covid-daywise.csv','italy-covid-daywise.csv')
covid_df = pd.read_csv('italy-covid-daywise.csv')
covid_df
```

While we typically use a library like matplotlib or seaborn plot graphs with jupyter notebook, Pandas data frames and series also provide a handy .plot method for quick and easy plotting.

Let's plot a line graph showing how the number of daily cases varies over time using the plot method of a pandas series.

```python
covid_df.new_cases.plot()
```

While this plot shows the overall trend, it's hard to tell where the peak occured, as there are no dates on the X axis. We can use the date column as the index for the data frame to adress this issue

```python
covid_df.set_index('date', inplace=True)
```

```python
covid_df
```

**Now let's plot the graph**

```python
covid_df.new_cases.plot()
```

We can observe that instead of serial numbers, dates were occured on x axis

```python
#Now we can retrieve data using dates
covid_df.loc['2020-09-01']
```

```python
 # let's plot new cases and new deaths per day as a combined graphs
 covid_df.new_cases.plot()
 covid_df.new_deaths.plot()
```

the comparision between new deaths and new cases will be easy now

You can give title to the graph

```python
covid_df.new_cases.plot(title='New Cases in Italy')
```

```python
#we can change the line graphs to bar graphs
covid_df.new_cases.plot(kind='bar')
```

```python
covid_df.new_deaths.plot(kind='bar')
```

use bar graphs if you have smaller number of data on x axis
